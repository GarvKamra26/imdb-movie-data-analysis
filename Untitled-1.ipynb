# Exploratory Data Analysis on Movie Data
  Goals:
  1. Find average ratings of each genre and predict which is the most liked genre

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('imdb_top_movies_1980_2026.csv')
df['genres_list'] = df['genres'].str.split(',')
df['genres_list']
```

```python
df_exploded = df.explode('genres_list')
df_exploded.head()
```

```python
genre_averages = df_exploded.groupby('genres_list')['average_rating'].mean().round(3)
genre_averages_df = genre_averages.reset_index(name='average_rating')
genre_averages
```

```python
plt.figure(figsize=(24, 6))
plt.bar(genre_averages_df['genres_list'], genre_averages_df['average_rating'])
plt.show()
```

From this we can say that News is the most liked genre
