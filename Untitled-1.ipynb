# Exploratory Data Analysis on Movie Data
  Goals:
  1. Find average ratings of each genre and predict which is the most liked genre
  2. Finding out if movie length gotten longer over time

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Making Dataframe of the csv file
df = pd.read_csv('imdb_top_movies_1980_2026.csv')
df['genres_list'] = df['genres'].str.split(',')
```

```python
# Exploding dataframe so that each genre has its own row
df_exploded = df.explode('genres_list')
```

```python
# Calculating the average rating for each genre
genre_avgs = df_exploded.groupby('genres_list')['average_rating'].mean().round(3)
genre_avgs_df = genre_avgs.reset_index(name='average_rating')
```

```python
# Visualizing the average ratings for each genre
plt.figure(figsize=(24, 6))
plt.bar(genre_avgs_df['genres_list'], genre_avgs_df['average_rating'])
plt.xlabel('Genres')
plt.ylabel('Average Rating')
plt.title('Trend of Average Movie Ratings by Genre')
plt.xticks(rotation=45)
plt.show()
```

From this we can say that News is the most liked genre

```python
# Calculating average runtime for every year
runtime_avg = df.groupby('year')['runtime_minutes'].mean().round(3).reset_index(name='average_runtime')
```

```python
# Plotting correlation between average runtime and year
plt.plot(runtime_avg['year'], runtime_avg['average_runtime'])
plt.xlabel('Year')
plt.ylabel('Average Runtime (minutes)')
plt.title('Trend of Average Movie Runtime Over Time')
plt.show()
```

From this we can see that average runtime has infact increased over the years
